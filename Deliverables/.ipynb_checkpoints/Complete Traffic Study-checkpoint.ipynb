## <center>BEGIN DATA RETRIEVAL & PRELIMINARY CLEANING</center>

```python
# –––––––––––––––––––––––––––––––––––– * API CALLS * ––––––––––––––––––––––––––––––––––––  #
#Retrieve the data

start_date = date.date(year=2017, month=9, day=1)
end_date = date.date(year=2018, month=12, day=1)
period = date.timedelta(days=1)

data = []

date = start_date

while date < end_date:
    the_date = date.strftime('%Y-%m-%d')
    url = f"https://data.austintexas.gov/resource/r3af-2r8x.json?${api_key}$where=published_date between \'{the_date}T00:00:00\' and \'{the_date}T23:59:59\'"
    response = requests.get(url).json()
    data += response
    date += period
    
data
```

```python
# Converts the Data into a DataFrame
traffic_df = pd.DataFrame.from_dict(json_normalize(data), orient='columns')

# Save data as csv
traffic_df.to_csv(raw_dataset)

traffic_df.head()
```

```python
# ––––––––––––––––––––––––––––––––––––- * Check Missing Value * ––––––––––––––––––––––––––––––––––––- #
traffic_df.count()
```

```python
# ––––––––––––––––––––––––––––––––––––- * DROP NAS * ––––––––––––––––––––––––––––––––––––- #
raffic_df=traffic_df.dropna()
traffic_df.count()
```

```python
# ––––––––––––––––––––––––––––––––––––- * CATEGORIES? * ––––––––––––––––––––––––––––––––––––- #
traffic_df['issue_reported'].value_counts()
```

```python
# –––––––––––––––––––––––––––––––––––– * REORGANIZING CATEGORIES * ––––––––––––––––––––––––––––––––––––  #
# Clean up issue_reported category. Replace 'Accident/Collision' in a single category.

#ACCIDENT COLLISION
traffic_df['issue_reported'] = traffic_df['issue_reported'].replace({
    'Crash Urgent': 'Accident/Collision', 'Crash Service': 'Accident/Collision',
    'COLLISION': 'Accident/Collision', 'COLLISION WITH INJURY': 'Accident/Collision',
    'COLLISN/ LVNG SCN': 'Accident/Collision', 'COLLISN / FTSRA': 'Accident/Collision',
    'COLLISION/PRIVATE PROPERTY': 'Accident/Collision', 'COLLISN/ LVNG SCN': 'Accident/Collision',
    'AUTO/ PED': 'Accident/Collision','FLEET ACC/ INJURY': 'Accident/Collision',
    'TRAFFIC FATALITY': 'Accident/Collision'})

#TRAFFIC IMPEDIMENT
traffic_df['issue_reported'] = traffic_df['issue_reported'].replace({
    'zSTALLED VEHICLE': 'Traffic Impediment', 'BLOCKED DRIV/ HWY': 'Traffic Impediment',
    'LOOSE LIVESTOCK': 'Traffic Impediment'})


#TRAFFIC HAZARD
traffic_df['issue_reported'] = traffic_df['issue_reported'].replace({
    'TRFC HAZD/ DEBRIS': 'Traffic Hazard', 'N / HZRD TRFC VIOL': 'Traffic Hazard',
    'ICY ROADWAY': 'Traffic Hazard', 'HIGH WATER': 'Traffic Hazard',
    'VEHICLE FIRE': 'Traffic Hazard', 'HIGH WATER': 'Traffic Hazard'})

# Drop the values for 'BOAT ACCIDENT' since those are not of interest for the study.
traffic_df = traffic_df[traffic_df.issue_reported != 'BOAT ACCIDENT']

traffic_df['issue_reported'].value_counts()
```

```python
# ––––––––––––––––––––––––––––––––––––- * OUTPUT TO CSV * ––––––––––––––––––––––––––––––––––––- #
traffic_df.to_csv(dirty_dataset) #clean dataset to work with
```

# <center>INCIDENTS BY DAY</center>

```python
# ––––––––––––––––––––––––––––––––––––- * READ IN CSV * ––––––––––––––––––––––––––––––––––––- #
traffic_df = pd.read_csv(main_dataset)
traffic_df.head()
```

```python
# ––––––––––––––––––––––––––––––––––––- * NEW COLUMN 'DAY' = DAY OF WEEK * ––––––––––––––––––––––––––––––––––––- #
traffic_df['Date']=pd.to_datetime(traffic_df['Date'])
traffic_df['Day'] = traffic_df['Date'].dt.weekday_name
traffic_df.head()
```

```python
# Find the total number of incidents that occurred for each date and what day of the week it was
grouped_df = traffic_df.groupby(['Day','Date'])[['Issue Reported']].count()
grouped_df.head()
```

```python
# Use the unstack() method to pivot grouped_df
unstacked_df = grouped_df.unstack(level=0)

# Reorder the columns
unstacked_df = unstacked_df[[('Issue Reported', 'Monday'),
 ('Issue Reported', 'Tuesday'),
 ('Issue Reported', 'Wednesday'),
 ('Issue Reported', 'Thursday'),
 ('Issue Reported', 'Friday'),
 ('Issue Reported', 'Saturday'),
 ('Issue Reported', 'Sunday')]]
unstacked_df.head()
```

```python
# Find the average number of incidents for each day of the week
day_averages = unstacked_df.mean()
day_averages
```

```python
# ––––––––––––––––––––––––––––––––––––- * BAR CHART * ––––––––––––––––––––––––––––––––––––- #
x_axis = np.arange(len(day_averages))
sns.barplot(x_axis, day_averages)
#plt.bar(x_axis, day_averages, color='b', alpha=0.5, align="center")
# Tell matplotlib where we would like to place each of our x axis headers
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], rotation=0)
# Give chart a title and labels
plt.title("Average Number of Incidents by Day of the Week", fontsize=20)
plt.xlabel("Day of Week", fontsize=15)
plt.ylabel("Number of Incidents", fontsize=15)
# Save the plot and display it
plt.savefig("../Deliverables/Plot-Bar - Avg Issues (per Weekday).png")
plt.show()
```

```python
# ––––––––––––––––––––––––––––––––––––- * READ IN CSV * ––––––––––––––––––––––––––––––––––––- #
traffic_df = pd.read_csv(dirty_dataset)
traffic_df.head()
```

```python
# -------------------------------- HEAT MAP DATA AND PLOTTING -------------------------------------
# Creating a dataframe of the coordinates of each crash
traffic_lats = traffic_df['latitude']
traffic_lngs = traffic_df['longitude']
coords_df = pd.DataFrame({'latitude': traffic_lats,
                           'longitude': traffic_lngs}, dtype= float)

# Some values in the latitude column were missing their decimal point, so we had to add it:

coords_df.loc[coords_df['latitude'] == 29880130.0, 'latitude'] = 29.8801300
coords_df.loc[coords_df['latitude'] == 29804156.0, 'latitude'] = 29.8041560
coords_df.loc[coords_df['latitude'] == 29932409.0, 'latitude'] = 29.9324090

coords_df = coords_df[np.isfinite(coords_df['latitude'])] #Delete Nan Rows
# Show dataframe
coords_df.head()
```

```python
#coords_df[coords_df['longitude'].isnull()] #Check for NaN rows;
coords_df[coords_df['latitude'].isnull()]
```

```python
# Creating the base map
gmaps.configure(api_key= gkey)
fig4 = gmaps.figure()
traffic_coords = (30.2672, -97.7431)
fig4 = gmaps.figure(center=traffic_coords, zoom_level=12, map_type='ROADMAP')

# Adding the "heat" layer and zooming capabilities 
heatmap_layer = gmaps.heatmap_layer(coords_df) 
fig4.add_layer(heatmap_layer)
heatmap_layer.max_intensity = 30 #prevents ind. pts from disappearing when zooming in
heatmap_layer.point_radius = 12 #this needs to be a low value to counteract the intensity

#gmaps.figure()
fig4
```
