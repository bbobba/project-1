```python
# –––––––––––––––––––––––––––––––––- * READING & PREVIEWING CLEANED DATA * –––––––––––––––––––––––––––––––––-
traffic_df = pd.read_csv('Resources/traffic (main).csv')
traffic_df.head()
```

```python
# ––––––––––––––––––––––––––- * DF TO HOLD ALL ISSUES FOR ALL CATEGORIES PER DAY * ––––––––––––––––––––––––––-
dategroup = traffic_df.groupby('Date').count()
incidentsperday = dategroup['Issue Reported']
```

```python
# ––––––––––––––––––––––––––- * RETRIEVING DAYS W/ > 250 TOTAL ISSUES REPORTED * ––––––––––––––––––––––––––-
sortedtotalperday_df = incidentsperday.sort_values(ascending=False)
sortedtotalperday_df[0:13]

wrsttotday1x = sortedtotalperday_df.index[0]
wrsttotday1y = sortedtotalperday_df[0]

wrsttotday2x = sortedtotalperday_df.index[1]
wrsttotday2y = sortedtotalperday_df[1]

wrsttotday3x = sortedtotalperday_df.index[2]
wrsttotday3y = sortedtotalperday_df[2]

wrsttotday4x = sortedtotalperday_df.index[3]
wrsttotday4y = sortedtotalperday_df[3]

wrsttotday5x = sortedtotalperday_df.index[4]
wrsttotday5y = sortedtotalperday_df[4]

wrsttotday6x = sortedtotalperday_df.index[5]
wrsttotday6y = sortedtotalperday_df[5]

wrsttotday7x = sortedtotalperday_df.index[6]
wrsttotday7y = sortedtotalperday_df[6]

wrsttotday8x = sortedtotalperday_df.index[7]
wrsttotday8y = sortedtotalperday_df[7]

wrsttotday9x = sortedtotalperday_df.index[8]
wrsttotday9y = sortedtotalperday_df[8]

wrsttotday10x = sortedtotalperday_df.index[9]
wrsttotday10y = sortedtotalperday_df[9]

print(f'The 10 worst days of the year are ranked as:\n 1)  {wrsttotday1x}:{wrsttotday1y} , \n 2)  {wrsttotday2x}:{wrsttotday2y},\n 3)  {wrsttotday3x}:{wrsttotday3y},\n 4)  {wrsttotday4x}:{wrsttotday4y},\n 5)  {wrsttotday5x}:{wrsttotday5y},\n 6)  {wrsttotday6x}:{wrsttotday6y},\n 7)  {wrsttotday7x}:{wrsttotday7y},\n 8)  {wrsttotday8x}:{wrsttotday8y},\n 9)  {wrsttotday9x}:{wrsttotday9y},\n 10) {wrsttotday10x}:{wrsttotday10y}.')
```

```python
#Printing 14 days w/ least number of incidents; 
sortedtotalperday_df[-13:len(sortedtotalperday_df)]
```

```python
# ––––––––––––––––––––––––––- * PLOTTING TOTAL INCIDENTS PER DAY OVER TIME * ––––––––––––––––––––––––––-
incidentsperday.index = pd.to_datetime(incidentsperday.index)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%Y-%m-%d'))
ax.plot(incidentsperday.index, incidentsperday)
plt.xticks(rotation=70)
plt.title("Total Issues per Day Over Time", fontsize=25)
plt.xlabel("Date", fontsize=20)
plt.ylabel("Total Issues Reported", fontsize=20)
plt.grid()
plt.legend(loc='best', fontsize=20)
# plt.plot([wrsttotday1x], [wrsttotday1y], marker='o', markersize=3, color="red")
# plt.plot([wrsttotday2x], [wrsttotday2y], marker='o', markersize=9, color="red")
# plt.plot([wrsttotday3x], [wrsttotday3y], marker='o', markersize=8, color="red")
# plt.plot([wrsttotday4x], [wrsttotday4y], marker='o', markersize=7, color="red")
# plt.plot([wrsttotday5x], [wrsttotday5y], marker='o', markersize=6, color="red")
# plt.plot([wrsttotday6x], [wrsttotday6y], marker='o', markersize=5, color="red")
# plt.plot([wrsttotday7x], [wrsttotday7y], marker='o', markersize=4, color="red")
# plt.plot([wrsttotday8x], [wrsttotday8y], marker='o', markersize=3, color="red")
# plt.plot([wrsttotday9x], [wrsttotday9y], marker='o', markersize=2, color="red")
# plt.plot([wrsttotday10x], [wrsttotday10y], marker='o', markersize=1, color="red")
plt.savefig('../Deliverables/Annual Line Graph (Totals per Day).png') 
plt.show()
```

```python
dated_traffic_df = traffic_df.drop(columns=['Latitude', 'Longitude', 'Address', 'Location', 'Date Time Published', 'Traffic Report ID', 'Traffic Report Status (TRS)', 'TRS Date Time', 'Time'])
category_df=dated_traffic_df.groupby(['Date','Issue Reported']).count()
category_df.head()
```

```python
category_df.reset_index()
issue_r_df = category_df.pivot_table(index='Date',columns='Issue Reported', values='Unnamed: 0')
issue_r_df.head()
```

```python
# –––––––––––––––––––––––––––––- * RETRIEVING DAYS W/ > 150 ISSUES REPORTED * –––––––––––––––––––––––––––––-
sortedcollisionperday_df1 = sortedcollisionperday_df = issue_r_df['Accident/Collision'].sort_values(ascending=False)
sortedcollisionperday_df = sortedcollisionperday_df[0:13]

wrstday1x = sortedcollisionperday_df.index[0]
wrstday1y = sortedcollisionperday_df[0]

wrstday2x = sortedcollisionperday_df.index[1]
wrstday2y = sortedcollisionperday_df[1]

wrstday3x = sortedcollisionperday_df.index[2]
wrstday3y = sortedcollisionperday_df[2]

wrstday4x = sortedcollisionperday_df.index[3]
wrstday4y = sortedcollisionperday_df[3]

wrstday5x = sortedcollisionperday_df.index[4]
wrstday5y = sortedcollisionperday_df[4]

wrstday6x = sortedcollisionperday_df.index[5]
wrstday6y = sortedcollisionperday_df[5]

wrstday7x = sortedcollisionperday_df.index[6]
wrstday7y = sortedcollisionperday_df[6]

wrstday8x = sortedcollisionperday_df.index[7]
wrstday8y = sortedcollisionperday_df[7]

wrstday9x = sortedcollisionperday_df.index[8]
wrstday9y = sortedcollisionperday_df[8]

wrstday10x = sortedcollisionperday_df.index[9]
wrstday10y = sortedcollisionperday_df[9]

print(f'The 10 worst days of the year are ranked as:\n 1)  {wrstday1x}:{wrstday1y} , \n 2)  {wrstday2x}:{wrstday2y},\n 3)  {wrstday3x}:{wrstday3y},\n 4)  {wrstday4x}:{wrstday4y},\n 5)  {wrstday5x}:{wrstday5y},\n 6)  {wrstday6x}:{wrstday6y},\n 7)  {wrstday7x}:{wrstday7y},\n 8)  {wrstday8x}:{wrstday8y},\n 9)  {wrstday9x}:{wrstday9y},\n 10) {wrstday10x}:{wrstday10y}.')
```

```python
issue_r_df.index = pd.to_datetime(issue_r_df.index)

fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.xaxis.set_major_locator(dates.MonthLocator())
ax1.xaxis.set_minor_locator(dates.MonthLocator())
ax1.xaxis.set_major_formatter(dates.DateFormatter('%Y-%m-%d'))
ax1.plot(issue_r_df.index, issue_r_df['Accident/Collision'])
ax1.plot(issue_r_df.index, issue_r_df['Traffic Hazard'])
ax1.plot(issue_r_df.index, issue_r_df['Traffic Impediment'])
plt.title("Total Issues per Category Over Time", fontsize=20)
plt.xlabel("Date", fontsize=15)
plt.ylabel("Total Issues Reported", fontsize=15)
plt.legend(loc='best', fontsize=15)
plt.xticks(rotation=70)

# plt.plot([wrstday1x], [wrstday1y], marker='o', markersize=10, color="red")
# plt.plot([wrstday2x], [wrstday2y], marker='o', markersize=9, color="red")
# plt.plot([wrstday3x], [wrstday3y], marker='o', markersize=9, color="red")
# plt.plot([wrstday4x], [wrstday4y], marker='o', markersize=8, color="red")
# plt.plot([wrstday5x], [wrstday5y], marker='o', markersize=8, color="red")
# plt.plot([wrstday6x], [wrstday6y], marker='o', markersize=7, color="red")
# plt.plot([wrstday7x], [wrstday7y], marker='o', markersize=6, color="red")
# plt.plot([wrstday8x], [wrstday8y], marker='o', markersize=6, color="red")
# plt.plot([wrstday9x], [wrstday9y], marker='o', markersize=6, color="red")
# plt.plot([wrstday10x], [wrstday10y], marker='o', markersize=4, color="red")
plt.grid()
plt.savefig('../Deliverables/Annual Line Graph (Totals per Category).png') 
plt.show()
```
